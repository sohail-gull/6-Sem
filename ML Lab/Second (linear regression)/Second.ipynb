```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv('Salary_Data.csv')
data1.head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x='YearsExperience',y='Salary',data=data1,marker='*',color='red')
plt.title('Years Experience vs Salary')
plt.xlabel('Experience(in years)')
plt.ylabel('Salary(in INR)')
plt.show()
```

```python
X=data1.drop('Salary',axis=1)
X.head(5)
```

```python
y=data1[['Salary']]
y.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X=StandardScaler()
sc_y=StandardScaler()
```

```python
X_train=sc_X.fit_transform(X_train)
```

```python
y_train=sc_y.fit_transform(y_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred_train=regressor.predict(X_train)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_train,y_train)
plt.plot(X_train,y_pred_train,color='green')
plt.title('Years Experience vs Salary')
plt.xlabel('Experience(in years)')
plt.ylabel('Salary(in INR)')
plt.show()
```

```python
X_test=sc_X.fit_transform(X_test)
y_test=sc_y.fit_transform(y_test)
```

```python
y_pred_test=regressor.predict(X_test)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_test,y_test)
plt.plot(X_test,y_pred_test,color='green')
plt.title('Years Experience vs Salary(on test data)')
plt.xlabel('Experience(in years)')
plt.ylabel('Salary(in INR)')
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred_test)
score
```
